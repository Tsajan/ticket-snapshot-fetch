```python
import pandas as pd
import numpy as np
```

```python
testDF = pd.read_csv('./ticketMapping.csv', header=0)
testDF
```

```python
testDF = testDF.drop(['TokenId'], axis=1)
testDF
```

```python
testDF['totalGoldTickets'] = testDF.groupby(["WalletAddr"])["isGold"].transform("sum")
testDF['totalSilverTickets'] = testDF.groupby(["WalletAddr"])["isSilver"].transform("sum")
testDF['totalBronzeTickets'] = testDF.groupby(["WalletAddr"])["isBronze"].transform("sum")
```

```python
testDF = testDF.drop(['isGold', 'isSilver', 'isBronze'], axis=1)
testDF = testDF.drop_duplicates()
```

```python
testDF
```

```python
testDF.to_csv('ticketSnapShot.csv', index=False)
```
