```python
import pandas as pd
import numpy as np
```

```python
binanceDF = pd.read_csv('./ticketMapping20230104.csv', header=0)
binanceDF
```

```python
binanceDF = binanceDF.drop(['TokenId'], axis=1)
binanceDF
```

```python
binanceDF['bGoldTickets'] = binanceDF.groupby(["Wallet"])["isGold"].transform("sum")
binanceDF['bSilverTickets'] = binanceDF.groupby(["Wallet"])["isSilver"].transform("sum")
binanceDF['bBronzeTickets'] = binanceDF.groupby(["Wallet"])["isBronze"].transform("sum")
```

```python
binanceDF = binanceDF.drop(['isGold', 'isSilver', 'isBronze'], axis=1)
binanceDF = binanceDF.drop_duplicates()
```

```python
binanceDF
```

```python
binanceDF.to_csv('ticketBinance20230104.csv', index=False)
```

```python
polygonTicketsDF = pd.read_csv('./ticketPolygon20230104.csv', header = 0)
polygonTicketsDF = polygonTicketsDF.drop(['pTotalCount'], axis=1)
```

```python
polygonTicketsDF.sum()
```

```python
mergedDF = binanceDF.merge(polygonTicketsDF, on="Wallet", how="outer")
mergedDF = mergedDF.fillna(0)
mergedDF
```

```python
mergedDF['goldTickets'] = mergedDF['bGoldTickets'] + mergedDF['pGoldCount']
mergedDF['silverTickets'] = mergedDF['bSilverTickets'] + mergedDF['pSilverCount']
mergedDF['bronzeTickets'] = mergedDF['bBronzeTickets'] + mergedDF['pBronzeCount']
mergedDF
```

```python
mergedDF.to_csv('finalTicketSnapshot20230104.csv', index=False)
```
